```python
import numpy as np
import json
import matplotlib.pyplot as plt
import matplotlib as mpl
import matplotlib.lines as mlines

# Optional: Use a nice matplotlib style
plt.style.use("seaborn-v0_8-whitegrid")  # or "ggplot", "bmh", etc.

# Set font sizes globally
mpl.rcParams.update({
    "font.size": 12,
    "axes.titlesize": 14,
    "axes.labelsize": 12,
    "legend.fontsize": 11,
    "xtick.labelsize": 11,
    "ytick.labelsize": 11
})

# Use consistent colors or default matplotlib color cycle
colors = {
    'Fourier': '#1f77b4',    # blue
    'Radial': '#2ca02c',     # green
    'Indicator': '#d62728'   # red
}
```

```python
num_ks = [16,32,64,128,256]
for num_k in num_ks:
    divergence_fourier = np.load(rf"Divergence_Data/Fourier_{num_k}.npy")
    divergence_radial = np.load(rf"Divergence_Data/Radial_{num_k}.npy")
    divergence_Indicator = np.load(rf"Divergence_Data/Finite_{num_k}.npy")
    plt.plot(divergence_fourier, label = 'fourier')
    plt.plot(divergence_Indicator, label = 'Indicator')
    plt.plot(divergence_radial, label = "radial")
    plt.yscale("log", base=10)
    plt.legend()
    plt.title(f'Num_basis = {num_k}')
    plt.show()
```

```python
num_ks = [16,32,64,128,256]
after_100k_indicator = []
after_100k_radial = []
after_100k_fourier = []

idx = 0
for num_k in num_ks:
    divergence_fourier = np.load(rf"Divergence_Data/Fourier_{num_k}.npy")
    divergence_radial = np.load(rf"Divergence_Data/Radial_{num_k}.npy")
    divergence_Indicator = np.load(rf"Divergence_Data/Finite_{num_k}.npy")
    after_100k_indicator.append(divergence_Indicator[idx])
    after_100k_radial.append(divergence_radial[idx])
    after_100k_fourier.append(divergence_fourier[idx])
    
plt.plot(num_ks, after_100k_fourier, label = 'fourier')
plt.plot(num_ks, after_100k_indicator, label = 'Indicator')
plt.plot(num_ks, after_100k_radial, label = "radial")
plt.xscale("log", base=2)
plt.yscale("log", base=10)
plt.xlabel('Number of basis per dim')
plt.ylabel('KLD')
plt.legend()
plt.title(f'Performance after {(idx + 1) * 10}k steps')
plt.show()
```
